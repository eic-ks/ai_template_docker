```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

X, y = load_iris(return_X_y=True)

clf = RandomForestClassifier(random_state=42)
clf.fit(X, y)

print(f"Accuracy: {clf.score(X, y):.3f}")
```
